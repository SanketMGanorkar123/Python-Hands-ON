```python
import pandas as pd
import numpy as np

from sklearn.impute import KNNImputer
```

```python
df=pd.DataFrame({"P": [40,45,43,np.nan,41,48,50],
                  "Q" : [10,28,13,16,20,17,15],
                  "R": [1,5,1,2,8,3,9]})
df
```

```python
df["P"].mean()
```

Nan Eucledian Distance = sqrt[weight*distance]

weight = (Total no of coordinates) / (No of Present Coordinates)


```python
knn_imputer = KNNImputer(n_neighbors=3)
array1 = knn_imputer.fit_transform(df)
df1 = pd.DataFrame(array1,columns=df.columns)
df1
```

```python
df=pd.DataFrame({"P": [40,45,43,np.nan,41,48,50],
                  "Q" : [10,28,13,16,20,np.nan,15],
                  "R": [1,5,1,2,8,3,9],
                  "S": [1,0,2,np.nan,np.nan,2,0]})
df
```

```python
knn_imputer = KNNImputer(n_neighbors=1)
array1 = knn_imputer.fit_transform(df)
df1=pd.DataFrame(array1,columns=df.columns)
df1
```
