```python
from abc import ABC, abstractmethod
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return 3.14*self.radius*self.radius
circle=Circle(5)
circle.area()
```

```python
class Vehicle():
    def __init__(self,make,model,year):
        self.make =make
        self.model = model
        self.year = year
        self.is_running=False
    def start(self):
        self.is_running =True
        print(f"{self.make} {self.model} engine started")
    def stop(self):
        self.is_running =False
        print(f"{self.make} {self.model} engine stopped")
class Car(Vehicle):
    def __init__(self,make,model,year,num_doors):
        super().__init__(make,model,year)
        self.num_doors = num_doors
    def open_trunk(self):
        print(f"{self.make} {self.model} trunk opened")
class Truck(Vehicle):
    def __init__(self,make,model,year,max_load):
        super().__init__(make,model,year)
        self.max_load = max_load
    def load_cargo(self):
        print(f"{self.make} {self.model} cargo loaded")
class Motorcycle(Vehicle):
    def __init__(self,make,model,year,has_sidecar):
        super().__init__(make,model,year)
        self.has_sidecar = has_sidecar
    def wheelie(self):
        print(f"{self.make} {self.model} wheelie started")
car = Car("Toyota", "Camry", 2023, 4)
truck = Truck("Ford", "F-150", 2023, 2000)
motorcycle = Motorcycle("Harley-Davidson", "Sportster", 2023, True)
car.start()
car.open_trunk()
```

```python
class Content():
    def __init__(self,title,duration):
        self.title = title
        self.duration = duration
    def play(self):
        print(f"Playing {self.title}-{self.duration} minutes")
class User():
    def __init__(self,username):
        self.username = username
    def login(self):
        print(f"user {self.username} has logged in")
class PremiumUser(User):
    def __init__(self,username,subscription_level):
        super().__init__(username)
        self.subscription_level = subscription_level
    def access_premium_content(self):
        print(f"Premium user{self.username} has access to premium content")
class Movie(Content):
    def __init__(self,title,duration,genre):
        super().__init__(title,duration)
        self.genre = genre
class TVShow(Content):
    def __init__(self,title,duration,num_episodes):
        super().__init__(title,duration)
        self.num_episodes = num_episodes
class Documetray(Content):
    def __init__(self,title,duration,topic):
        super().__init__(title,duration)
        self.topic = topic
class WatchlistItem():
    def __init__(self,user,content):
        self.user = user
        self.content = content
    def add_to_watchlist(self):
        print(f"{self.user.username} added {self.content.title} to their watchlist")
class PremiumMovie(Movie,PremiumUser):
    def __init__(self,title,duration,genre,username,subscription_level):
        Movie.__init__(self,title,duration,genre)
        PremiumUser.__init__(self,username,subscription_level)
    def play(self):
        super(Movie,self).play()
        print("This is a premium movie")
user1 = User("user123")
premium_user1 = PremiumUser("premium_user456", "Gold")
movie1 = Movie("Inception", 150, "Science Fiction")
premium_movie1 = PremiumMovie("The Dark Knight", 180, "Action", "premium_user456", "Platinum")
user1.login()
```

```python
class Accessitem():
    def __init__(self,start_num,end_num):
        self.start_num = start_num
        self.end_num = end_num
    def __iter__(self):
        self.a=self.start_num
        return self
    def __next__(self):
        if self.a<=self.end_num:
            x=self.a
            self.a+=1
            return x
        else:
            raise StopIteration
obj=Accessitem(10,20)
my_iter=iter(obj)
for i in obj:
    print(i)
```

```python
class ExponentOfNumber():
    def __init__(self,number,max_power):
        self.number = number
        self.max_power = max_power
    def __iter__(self):
        self.n=1
        return self
    def __next__(self):
        if self.n<=self.max_power:
            result=self.number**self.max_power
            self.n+=1
            return result
        else:
            raise StopIteration
obj=ExponentOfNumber(5,3)
obj1=iter(obj)
next(obj1)
next(obj1)
```

```python
import math
class Circle():
    def __init__(self,radius):
        self.radius = radius
    def area(self):
        ar=math.pi*self.radius*self.radius
        print("Area of circle",ar)
class Square():
    def __init__(self,side):
        self.side = side
    def area(self):
        ar=self.side**2
        print("Area of square",ar)
class Rectangle():
    def __init__(self,l,b):
        self.l = l
        self.b = b
    def area(self):
        ar=self.l*self.b
        print("Area of rectangle",ar)
def execute_functions():
    c1=Circle(5)
    s1=Square(5)
    r1=Rectangle(2,4)
    for obj in [c1,s1,r1]:
        obj.area()
```
