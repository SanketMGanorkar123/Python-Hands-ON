> How Image is Generated?
* Image is collection of Pixels.
* Pixel is a Grid like Structure.
* Resolution is a count of pixels.
* Pixel has a concept at backend it is known as take any three primary color (Red, Green, Blue) change their color intensity(Brightness intensity) , by doing this we can form any color.
> Color image is in the format of RGB format.(RGB scale)
* It has 3 primary color (Red, Green, Blue)
* Range is 0 to 255.
* If we take 0 it means No Brightness(Black Color).(Red = 0 , Green = 0 , Blue = 0)
* If we take 255 it means Full Brightness(White color). (Red = 255 , Green = 255 , Blue = 255)
* RGB scale is used to form color images.
> Gray scale (Black and White)
* Primary color is Black.
* Range is 0 to 255.
* If we take 0 it means No Brightness(Black Color).(Black = 0)
* If we take 255 it means Full Brightness(White Color).(Black = 255)
* In between the range 0 to 255 we can make Gray color.
> Color depth / depth of color
* How many primary colors are used to form any image.
* Color image >> RGB scale >> 3 primary colors 
* So color depth is 3.
* Black and White images?
* Gray scale >> 1 primary color  
* So color depth is 1.

> How CNN works?
* In CNN we add two additional layers
> 1st layer Convolutional layer.
* It is used to extract features from the images.
> How Convolutional layer works?
* Suppose we have 4*4*1 (Black and white image) original image.
* Suppose we have 3*3 filter.
* We can apply filter on the original image to create as many different filter images.These filter images are different from each other.
* We can apply multiplication logic or Division logic on these filter images to create the resultant output.
* Suppose we have 2 images as original image and if we apply 5 filters each on these two images then we will get 10 images as output.
* IF our filter is of 3*3 matrix and our original image is of 4*4 matrix then in order to convert the filter image to 4*4 matrix we need to perform  Zero padding.(Adding zero for the blank matrix)
* Applying filter to the original image creates variation of the original image so that while we train our model we can have different variation to train our model with ,more the variation more nicely our model will get trained and the accuracy of the model will get increased.
> Rules for filter 
* We need to select the filter matrix as odd number to avoid overlapping(3*3 , 5*5 , 7*7 , 9*9).
* Filter image matrix depends upon size of the original image.
* If original image size is bigger then the size of the filter image will be bigger and if the original image size is smaller then the size of the filter image will be smaller.
> Stride 
* By how many pixels filter get moved to form new features.
* 
> 2nd layer Pooling layer.
* It is used to reduce size of extracted features.
* We have 3 types 
1. Max pooling (When image qaulity is bad)
2. Average pooling  (When image qaulity is good)
3. Sum pooling ( We dont use this quite often)
* So when we have good quality of images we use average pooling.So that we can bring some qaulity pixel values in to  our consideration to build a robust model.
* When we have bad quality of images that time we use max pooling.So that we can build our model on godd quality pixel values.
* 
> Flow of CNN 
1. Image data 
2. We will apply convolutional layer on Image data.
3. To Extract features
4. Then we will apply pooling layer to reduce the size of extracted features.
5. Then we will flatten our data that is we are converting mulidimensional matrix into one dimensional array.
6. Then we will apply fully connected layer(ANN).
7. At the end we will get classification.
* Convolutional layer works in sets (1 Convolutional layer and 1 pooling layer)
> What do you mean by flattening of data in CNN ? 
* We apply ANN in case of High complex structered data.
* we are converting mulidimensional array into one dimensional array.
* That is we are converting it into a structured data.
