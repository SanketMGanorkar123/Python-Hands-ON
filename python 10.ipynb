```python
for x_val in arr2:
    z_score=(x_val-mean)/std
    z_score=np.abs(z_score)
    print(f"{x_val} Z score is:",z_score)
```

```python
outliers_list=[]
threshold=3
for x_val in arr2:
    z_score=(x_val-mean)/std
    z_score=np.abs(z_score)
    print(f"{x_val} Z score is:",z_score)

    if z_score>threshold:
        outliers_list.append(x_val)
print("Outliers are:",outliers_list)
```

```python
from scipy import stats
```

```python
stats.zscore(arr2)
```

```python
z_score=np.abs(stats.zscore(arr2))
print(z_score)
threshold=3
outliers_list=[]
for i in z_score:
    if i>threshold:
        outliers_list.append(i)
print("Outliers list:", outliers_list)

    
```

```python
z_score=np.abs(stats.zscore(arr2))
print(z_score)
threshold=3
outliers_list=[]
for i in range(len(z_score)):
    if z_score[i]>threshold:
        outliers_list.append(arr2[i])
print("Outliers list:" ,outliers_list )
```
